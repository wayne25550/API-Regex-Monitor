```python
import requests
import re
import json

#從網路API抓取JSON資料
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)

if response.status_code == 200:
    users = response.json() # 將抓到的內容轉為 Python 列表
    
    #定義(Regex)，找出所有.biz的Email
    email_pattern = r"[\w.+-]+@[\w.-]+\.biz"
    
    #設定電話格式：包含數字、點、橫線、空格或分機號(x)，長度至少10碼
    phone_pattern = r"[\d.x -]{10,}"

    print(f"--- 成功抓取 {len(users)} 筆資料，開始進行格式審計 --- \n")

    for user in users:
        email = user['email']
        phone = user['phone']
        name = user['name']

        #使用Regex 進行比對
        is_biz_email = re.search(email_pattern, email)
        
        if is_biz_email:
            print(f"找到目標用戶:{name}")
            print(f"信箱:{email} (符合 .biz 格式)")
            
            # 順便清洗電話資料，只保留純數字 (示範 re.sub)
            clean_phone = re.sub(r"\D", "", phone) #將「非數字」的字元，全部取代成空字串
            print(f"原始電話:{phone}")
            print(f"清洗後純數字:{clean_phone}")
            print("-" * 30)

else:
    print("無法連線至 API")
```
